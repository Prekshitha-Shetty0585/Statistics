```python
#Data visualization
#histogram,boxplot,barplot for categorical colums
```

```python
# Identify numerical columns
numerical_columns = Pre.select_dtypes(include=['float64', 'int64']).columns
print(numerical_columns)
```

```python
# Plot histograms for each numerical column
for col in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(Pre[col], bins=30, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
print(Pre['Volume'].skew())
print(Pre['Volume'].kurtosis())
```

```python
# Create boxplots for numerical variables
for col in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=Pre[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

```python
#### Bar Chart Analysis for Categorical Column
```

```python
# Identify categorical columns
categorical_columns = Pre.select_dtypes(include=['object', 'category']).columns
print(categorical_columns)

# Plot bar charts for each categorical column
for col in categorical_columns:
    plt.figure(figsize=(10, 5))
    Pre[col].value_counts().plot(kind='bar')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
# Standardize numerical columns
standardized = Pre[numerical_columns].apply(lambda x: (x - x.mean()) / x.std())

# Before and after comparison
print("Original Data:\n", Pre[numerical_columns].head())
print("\nStandardized Data:\n", standardized.head())

# Visualizing the standardized data
for col in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(standardized[col], bins=30, kde=True)
    plt.title(f'Standardized Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
## Conversion of Categorical Data into Dummy Variables
#Objective: To transform categorical variables into a format that can be provided to ML algorithms
```

```python
# Apply one-hot encoding
df_encoded = pd.get_dummies(Pre, columns=categorical_columns, drop_first=True)

# Display a portion of the transformed dataset
print(df_encoded.head())
```
